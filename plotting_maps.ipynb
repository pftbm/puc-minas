```python
import folium
import pandas as pd
```

```python
data_aptos = pd.read_excel('data_aptos_final.xlsx')
```

MARCADOR COMUM

```python
map = folium.Map(location=[-8.1223044,-34.9022416], zoom_start=14)
locations = data_aptos[['lat_bing', 'lon_bing']]
locationlist = locations.values.tolist()

for point in range(0, len(data_aptos)):
    folium.Marker(locationlist[point]).add_to(map)
map
```

MARCADOR "CIRCLE MARKER"

```python
this_map = folium.Map(prefer_canvas=True)

color_marker = 'navy'
r_marker = 2
w_marker = 2

def plotDot(point):
    '''input: series that contains a numeric named latitude and a numeric named longitude
    this function creates a CircleMarker and adds it to your this_map'''
    folium.CircleMarker(location=[point.lat_bing, point.lon_bing],
                        radius=r_marker,
                        weight=w_marker, color=color_marker).add_to(this_map)

#use df.apply(,axis=1) to "iterate" through every row in your dataframe
data_aptos.apply(plotDot, axis = 1)


#Set the zoom to the maximum possible
this_map.fit_bounds(this_map.get_bounds())

#Save the map to an HTML file
this_map.save('simple_dot_plot_navy.html')

this_map
```

HEAT MAP + CIRCLE MARKER
