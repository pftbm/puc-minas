```python
first = True
inicio = 1908
anos_excesso = [1939, 1945, 1953, 1956, 1968, 1969, 1971, 1972, 1973, 1982, 1983, 1997, 1998, 1999]
anos_ultra_excesso = [1939, 1982, 1997, 1999]
new_data = []

for ano in range(inicio, 2019): 
    
    if ano not in anos_excesso:
        url = "https://esigportal2.recife.pe.gov.br/arcgis/rest/services/Financas/BASES_LOTES_FEICAO/MapServer/0/query?where=ANCONSTR+%3D+"+str(ano)+"&text=&objectIds=&time=&geometry=&geometryType=esriGeometryEnvelope&inSR=&spatialRel=esriSpatialRelIntersects&relationParam=&outFields=OBJECTID%2C+ENDNUMERO%2C+DB2GSE.ST_Area%28SHAPE%29%2C+AREALOTE%2C+AREATOTALCONSTRUIDA%2C+QTDPAVIMENTOS%2C+TIPOEMPREENDIMENTO%2C+TESTADAPRINCIPAL%2C+DSQFL%2C+NMEDIFICACAO%2C+ANCONSTR%2C+QTDMULTIPLAS%2C+QTDUNHAB%2C+NMENDCOMP%2C+EFTUTZDESC&returnGeometry=false&returnTrueCurves=false&maxAllowableOffset=&geometryPrecision=&outSR=&having=&returnIdsOnly=false&returnCountOnly=false&orderByFields=&groupByFieldsForStatistics=&outStatistics=&returnZ=false&returnM=false&gdbVersion=&historicMoment=&returnDistinctValues=false&resultOffset=&resultRecordCount=&queryByDistance=&returnExtentOnly=false&datumTransformation=&parameterValues=&rangeValues=&quantizationParameters=true&featureEncoding=esriDefault&f=pjson"
        response = urlopen(url)

        if first:
            data_json = json.loads(response.read(), strict=False)
            first = False
            print('ano :', ano, 'elementos: ', len(data_json['features']))            
        else:
            new_data = json.loads(response.read(), strict=False)['features']
            data_json['features'].extend(new_data)        
            print('ano :', ano, 'elementos: ', len(new_data))
            if len(new_data) > 1000:
                print("#################################### maior que 1000!")

    else:
        min = 0
        faixas = 11
        offset = 50
        
        if ano in anos_ultra_excesso:                        
            if ano == 1982:
                offset = offset // 25
                faixas = faixas * 50
            else:
                offset = offset // 5
                faixas = faixas * 10
        
        max = min + offset
        
        for i in range(1, faixas):
            
            if i == (faixas - 1):
                max = 10000000
            
            url = "https://esigportal2.recife.pe.gov.br/arcgis/rest/services/Financas/BASES_LOTES_FEICAO/MapServer/0/query?where=ANCONSTR+%3D" + str(ano) + "++AND+DB2GSE.ST_Area(SHAPE)+%3E%3D+"  +str(min)+  "+AND+DB2GSE.ST_Area(SHAPE)+%3C+"  +str(max)+  "&text=&objectIds=&time=&geometry=&geometryType=esriGeometryEnvelope&inSR=&spatialRel=esriSpatialRelIntersects&relationParam=&outFields=OBJECTID%2C+ENDNUMERO%2C+DB2GSE.ST_Area%28SHAPE%29%2C+AREALOTE%2C+AREATOTALCONSTRUIDA%2C+QTDPAVIMENTOS%2C+TIPOEMPREENDIMENTO%2C+TESTADAPRINCIPAL%2C+DSQFL%2C+NMEDIFICACAO%2C+ANCONSTR%2C+QTDMULTIPLAS%2C+QTDUNHAB%2C+NMENDCOMP%2C+EFTUTZDESC&returnGeometry=false&returnTrueCurves=false&maxAllowableOffset=&geometryPrecision=&outSR=&having=&returnIdsOnly=false&returnCountOnly=false&orderByFields=&groupByFieldsForStatistics=&outStatistics=&returnZ=false&returnM=false&gdbVersion=&historicMoment=&returnDistinctValues=false&resultOffset=&resultRecordCount=&queryByDistance=&returnExtentOnly=false&datumTransformation=&parameterValues=&rangeValues=&quantizationParameters=&featureEncoding=esriDefault&f=pjson"
            response = urlopen(url)
            
            min = min + offset
            max = min + offset
            
            new_data = json.loads(response.read(), strict=False)['features']
            data_json['features'].extend(new_data)
                
           # if len(new_data) == 1000:
           #     print("Ano de ", ano, "possui mais de ", len(new_data), "dados.")
            
            print('ano: ', ano, ' faixa: ', i, ' elementos:', len(new_data))
            if len(new_data) > 1000:
                print("###################################3 maior que 1000!")
        
        
#data_json
len(data_json['features'])

```

```python

df = pd.json_normalize(data_json['features'])
bairros = []
ruas = []

for data in df['attributes.NMENDCOMP']:
    ruas.append(str.split(data, sep = ',')[0].strip())
    bairros.append(str.split(data, sep = ',')[2].strip())
    

bairros = pd.DataFrame(bairros, columns= ['Bairro'])

df['Bairro'] = bairros
df['Rua'] = ruas
df.rename(columns = {'attributes.OBJECTID': 'ID', 'attributes.DB2GSE.ST_Area(SHAPE)': 'Area da Forma',
                     'attributes.TIPOEMPREENDIMENTO':'Tipologia', 'attributes.QTDPAVIMENTOS':'Pavimentos',
                     'attributes.ENDNUMERO':'Numero', 'attributes.NMENDCOMP':'Endereco', 
                     'attributes.AREALOTE':'Area Lote', 'attributes.AREATOTALCONSTRUIDA':'Area Construida',
                     'attributes.NMEDIFICACAO':'Nome Edificacao', 'attributes.ANCONSTR':'Ano Construcao',
                     'attributes.TESTADAPRINCIPAL':'Testada Principal', 'attributes.DSQFL':'DSQFL',
                     'attributes.QTDMULTIPLAS':'Qtd Blocos', 'attributes.QTDUNHAB':'Qtd Unidades',
                     'attributes.EFTUTZDESC':'Tipo Residuo'
                     }, inplace = True)

df = df[["ID","DSQFL","Endereco","Rua","Numero", "Bairro", "Tipologia","Pavimentos", "Nome Edificacao", 
         "Area Lote","Area da Forma", "Area Construida","Testada Principal", "Ano Construcao", "Qtd Blocos",
         "Qtd Unidades", "Tipo Residuo"]]
#df[df['Bairro'] == 'Boa Viagem']
df
```
